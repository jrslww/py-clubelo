```python
import requests
import csv
from io import StringIO
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from unicodedata import normalize
from bs4 import BeautifulSoup
```

```python
def getRanking():
  day = datetime.date(datetime.now())
  return showRanking(day)

def showRanking(day):
  
  url = requests.get('http://api.clubelo.com/' + str(day))
  data = StringIO(url.text)
  df = pd.read_csv(data, sep=",")
  return df

def plotCountry(tag):
  df = getRanking()
  lst = df.loc[df['Country'] == str(tag)]
  k = []
  for i in lst['Club']:
    j = i.replace(' ','')
    k.append(j)

  return k


def getCountryElo(tag):
  fc_list = []
  for element in tag:
    fc_list.append(plotCountry(tag))
    print(fc_list)
  # return df["Elo"].mean()
  return fc_list

def plotClub(teamlist):
  fc_list = []
  data = StringIO()
  for element in teamlist:
    r = requests.get('http://api.clubelo.com/' + str(element))
    data = StringIO(r.text)
    df = pd.read_csv(data, sep=",")
    fc_list.append(df)
  return pd.concat(fc_list)
  # df.From = pd.to_datetime(df['From'])
  # df.To = pd.to_datetime(df['To'])

# print( getRanking()['Country'].unique())
df = getRanking()
# df =plotCountry(['UKR'])
df
# lst = [x.strip(' ') for x in lst]
# dataframe = plotClub(['DynamoKyiv', 'DinamoZagreb'])
# print (dataframe.Club.unique())    
```
